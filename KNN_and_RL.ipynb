## Algoritmo K-NN (K-Vizinhos mais Próximos)

# Algoritmo de aprendizado de máquina preguiçoso (lazy), ou seja, há a necessidade de se ter os dados toda a vez que usar o algoritmo.

O K-NN é baseado em técnicas de cálculo de distâncias. A distância é calculada com base nos vizinhos mais próximos.

```python
import pandas as pd
```

```python
data_set = pd.read_csv('/content/diabetes.csv')
```

```python
data_set.head(3)
```

```python
X = data_set.iloc[:, 0:8].values
y = data_set.iloc[:, 8].values
```

```python
X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_treino, X_teste, y_treino, y_teste = train_test_split(X, y)
```

```python
X_treino.shape, X_teste.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(X_treino, y_treino)
```

```python
predicoes = knn.predict(X_teste)
```

```python
predicoes
```

```python
y_teste == predicoes
```

```python
from yellowbrick.classifier import confusion_matrix
```

```python
cm = confusion_matrix(knn, X_teste, y_teste)
```

```python
from sklearn.metrics import accuracy_score
```

```python
ac = accuracy_score(y_teste, predicoes)
```

```python
print(ac)
```

## Regressão Logística

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_treino, y_treino)
```

```python
predicoes_lr = lr.predict(X_teste)
```

```python
cm_lr = confusion_matrix(lr, X_teste, y_teste)
```

```python
print(accuracy_score(y_teste, predicoes_lr))
```
