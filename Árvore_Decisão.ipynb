## Árvore de Decisão

Esta técnica serve tanto para classificação como para regressão. A estrutura básica é uma árvore binária. O processo de treinamento consiste em elaborar uma árvore contendo: nó raíz, nós internos (nós de decisão), nós folhas. Um nó folha é a resposta do modelo.



```python
import pandas as pd
```

Dataset de esportes:

Futebol
Volei
Atletismo

```python
esportes = ['futebol', 'volei', 'atletistmo']
```

```python
altura = [2.00, 1.50, 1.65, 1.70, 1.75, 2.09, 1.69, 1.85, 1.90, 1.58]
```

```python
idade = [13, 20, 35, 40, 22, 33, 16, 18, 65, 27]
```

```python
pratica = [esportes[1], esportes[0], esportes[2], esportes[2], esportes[1], esportes[1], esportes[0], esportes[0], esportes[1], esportes[2]]
```

```python
dados = pd.DataFrame({'Altura': altura,
                      'Idade': idade,
                      'Esporte que pratica': pratica})
```

```python
dados
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
arvore = DecisionTreeClassifier(criterion='entropy')
```

```python
X = dados.iloc[:, 0:2].values
y = dados.iloc[:, 2].values
```

```python
X.shape, y.shape
```

```python
arvore.fit(X, y)
```

```python
from sklearn import tree
```

```python
tree.plot_tree(arvore, filled=True, feature_names=['Altura', 'Idade'], class_names=esportes)
```

# Dataset de cancer de mama com Árvore de Decisão

```python
data_set = pd.read_csv('/content/breast-cancer.csv')
```

```python
data_set.head(5)
```

```python
# Saída (classes)
y = data_set.iloc[:, 1].values
```

```python
# Entrada de dados (atributos)
X = data_set.iloc[:, 2:32].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, random_state=20)
```

```python
X_treino.shape, y_treino.shape, X_teste.shape, y_teste.shape
```

```python
arvore_cancer = DecisionTreeClassifier(criterion='gini')
```

```python
arvore_cancer.fit(X_treino, y_treino)
```

```python
arvore_predicoes = arvore_cancer.predict(X_teste)
```

```python
data_set.index
```

```python
data_set.columns[2:-1]
```

```python
labels = data_set.columns[2:-1]
```

```python
tree.plot_tree(arvore_cancer, filled=True, feature_names=labels, class_names=arvore_cancer.classes_)
```

```python
from yellowbrick.classifier import ConfusionMatrix
```

```python
cm = ConfusionMatrix(arvore_cancer)
```

```python
cm.score(X_teste, y_teste)
```
